# Gantt chart

```python
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import glob
output_notebook()

p = figure(height=200, width=1000, tools='xwheel_zoom,xpan,reset,save')
colors = ['red', 'green', 'blue', 'orange',
          'purple', 'brown', 'pink', 'grey', 'black']

for i, csv in enumerate(sorted(glob.glob(f'{LOG_DIR}/callback*.csv'))):
    df = pd.read_csv(csv)
    df['time_ns'] -= df['time_ns'].min()
    source = ColumnDataSource(df)
    p.rect(x='time_ns', y=i, width='response_time_us', height=0.8,
           source=source, fill_color=colors[i], line_color=None)

hover = HoverTool()
hover.tooltips = [
    # ("Scheduling Policy", "@scheduling_policy"),
    ("Start Time [ns]", "@time_ns"),
    ("Response Time [us]", "@response_time_us")]
p.add_tools(hover)
show(p)
```
