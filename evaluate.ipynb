```python
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
output_notebook()

df = pd.read_csv('exec_time_logs/callback0.csv')
df['start_time_ms'] -= df['start_time_ms'].min()
source = ColumnDataSource(df)
p = figure()
p.rect(x='start_time_ms', y=0, width='execution_time_ms', height=0.8,
       source=source, fill_color="navy", line_color=None)

hover = HoverTool()
hover.tooltips = [("Start Time [ms]", "@start_time_ms"),
                  ("Execution Time [ms]", "@execution_time_ms")]
p.add_tools(hover)
show(p)
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import pandas as pd

# Activate inline plotting for the notebook
output_notebook()

# Prepare the data for plotting
data = [
    {"timestamp": 1705032327617, "duration": 96},
    {"timestamp": 1705032327716, "duration": 96},
    {"timestamp": 1705032327916, "duration": 96},
    {"timestamp": 1705032328116, "duration": 96},
    {"timestamp": 1705032328316, "duration": 96},
    {"timestamp": 1705032328516, "duration": 96},
    {"timestamp": 1705032328716, "duration": 96},
    {"timestamp": 1705032328916, "duration": 96},
    {"timestamp": 1705032329116, "duration": 96},
    {"timestamp": 1705032329316, "duration": 97}
]

df = pd.DataFrame(data)

# Convert timestamps to datetime
df['start_time'] = pd.to_datetime(df['timestamp'], unit='ms')
# df['end_time'] = df['start_time'] + pd.to_timedelta(df['duration'], unit='s')

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a figure
p = figure(x_axis_type="datetime", title="Gantt Chart Using Bokeh")

# Adding rects
# for i, data in enumerate(data):
p.rect(x='start_time', y=0, width='duration', height=0.8,
       source=source, fill_color="navy", line_color=None)

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Start", "@start_time{%F %T}"), ("Duration", "@duration s")]
hover.formatters = {'@start_time': 'datetime'}
p.add_tools(hover)

# Show the plot
show(p)
```
